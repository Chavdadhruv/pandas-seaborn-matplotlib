```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
dataset=pd.read_csv("auto-mpg.csv")
```

```python
fig,ax=plt.subplots()
dataset.boxplot(column=["mpg"],by="cylinders",ax=ax)
plt.show()
```

```python
fig,ax=plt.subplots()
dataset.boxplot(column=["mpg","acceleration"],widths=0.75,notch=True,ax=ax)
plt.show()
```

```python
import seaborn as sns
```

```python
sns.boxplot(x=dataset["cylinders"],y=dataset["mpg"],data=dataset)
plt.show()
```

```python
def remove_outline(dataset,col):
    q1=dataset[col].quantile(0.25)
    q3=dataset[col].quantile(0.75)
    iqr=q3-q1
    min=q1-1.5*iqr
    max=q3+1.5*iqr
    return dataset[(dataset[col]>=min)&(dataset[col]<=max)]
dataset=remove_outline(dataset,'mpg')
print(dataset)
```

```python
sns.boxplot(x=dataset["cylinders"],y=dataset["mpg"],data=dataset)
plt.show()
```

```python
months=np.arange(1,13)
sales=np.array([10,15,20,30,20,50,30,70,80,40,10,100])
expenses=np.array([5,10,15,20,25,30,35,40,45,50,55,60])
fig,ax=plt.subplots()
plt.plot(months,sales,color="green",label="sales")
plt.plot(months,expenses,color="red",label="expenses")
plt.fill_between(months,sales,expenses,where=sales>expenses,alpha=0.5,color="green",interpolate=True)
plt.fill_between(months,sales,expenses,where=expenses>sales,alpha=0.5,color="red",interpolate=True)
ax.legend()
plt.show()
fig.savefig("myplot.png",format='png')
```
