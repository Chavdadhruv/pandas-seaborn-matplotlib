```python
def remove_outline(dataset,col):
    q1=dataset[col].quantile(0.25)
    q3=dataset[col].quantile(0.75)
    iqr=q3-q1
    min=q1-1.5*iqr
    max=q3+1.5*iqr
    return dataset[(dataset[col]>=min)&(dataset[col]<=max)]
dataset=remove_outline(dataset,'acceleration')
print(dataset)
```

```python
import matplotlib.pyplot as plt
fig,ax=plt.subplots()
ax.boxplot([dataset['mpg'],dataset['acceleration']])
plt.show()
```

```python
data={'year':[2015,2016,2017,2018,2019,2020],
      'sales':[100,120,150,170,200,220],
      'expenses':[70,80,90,100,110,120]}
df=pd.DataFrame(data)
plt.stackplot(df['year'],df['sales'],df['expenses'],labels=['sales','expenses'])
plt.legend(loc='upper left')
plt.show()
```

```python
data={'year':[2015,2016,2017,2018,2019,2020],
      'sales':[100,50,150,100,150,50],
      'expenses':[70,80,90,100,110,120]}
df=pd.DataFrame(data)
plt.fill_between(df['year'],df['sales'],df['expenses'],alpha=0.5,color='red')
plt.show()
```
